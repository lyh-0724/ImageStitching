```python
from pylab import *
from numpy import *

np.seterr(divide='ignore', invalid='ignore')
from PIL import Image
import numpy as np
import math
import scipy.ndimage


def plane_sweep_ncc(im_l, im_r, start, steps, wid):
    """  """
    m, n = im_l.shape
    # 
    mean_l = np.zeros((m, n))
    mean_r = np.zeros((m, n))
    s = np.zeros((m, n))
    s_l = np.zeros((m, n))
    s_r = np.zeros((m, n))
    # 
    dmaps = np.zeros((m, n, steps))
    # 
    scipy.ndimage.uniform_filter(im_l, wid, mean_l)
    scipy.ndimage.uniform_filter(im_r, wid, mean_r)
    # 
    norm_l = im_l - mean_l
    norm_r = im_r - mean_r
    # 
    for displ in range(steps):
        # 
        scipy.ndimage.uniform_filter(np.roll(norm_l, -displ - start) * norm_r, wid, s)  # 
        scipy.ndimage.uniform_filter(np.roll(norm_l, -displ - start) * np.roll(norm_l, -displ - start), wid, s_l)
        scipy.ndimage.uniform_filter(norm_r * norm_r, wid, s_r)  # 
        #  ncc 
        dmaps[:, :, displ] = s / np.sqrt(s_l * s_r)
    # 
    return np.argmax(dmaps, axis=2)


def plane_sweep_gauss(im_l, im_r, start, steps, wid):
    """  """
    m, n = im_l.shape
    # 
    mean_l = np.zeros((m, n))
    mean_r = np.zeros((m, n))
    s = np.zeros((m, n))
    s_l = np.zeros((m, n))
    s_r = np.zeros((m, n))
    # 
    dmaps = np.zeros((m, n, steps))
    # 
    scipy.ndimage.gaussian_filter(im_l, wid, 0, mean_l)
    scipy.ndimage.gaussian_filter(im_r, wid, 0, mean_r)
    # 
    norm_l = im_l - mean_l
    norm_r = im_r - mean_r
    # 
    for displ in range(steps):
        # 
        scipy.ndimage.gaussian_filter(np.roll(norm_l, -displ - start) * norm_r, wid, 0, s)  # 
        scipy.ndimage.gaussian_filter(np.roll(norm_l, -displ - start) * np.roll(norm_l, -displ - start), wid, 0, s_l)
        scipy.ndimage.gaussian_filter(norm_r * norm_r, wid, 0, s_r)  # 
        #  ncc 
        dmaps[:, :, displ] = s / np.sqrt(s_l * s_r)
    # 
    return np.argmax(dmaps, axis=2)


""""""
im_l = np.array(Image.open('data/view1.png').convert('L'), 'f')
im_r = np.array(Image.open('data/view5.png').convert('L'), 'f')
figure()
gray()
subplot(221)
imshow(im_l)
subplot(222)
imshow(im_r)

# 
steps = 12
start = 4
# ncc 
wid = 13
res = plane_sweep_ncc(im_l, im_r, start, steps, wid)
wid2 = 5
gauss = plane_sweep_gauss(im_l, im_r, start, steps, wid2)
subplot(223)
imshow(res)
subplot(224)
imshow(gauss)
show()
```
